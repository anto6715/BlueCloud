```python
import xarray as xr
import numpy as np
import pyxpcm
from pyxpcm.models import pcm
```

```python
file_path = '/export/home1/agarciaj/datasets/global-reanalysis-phy-001-030-monthly_1598523953615.nc'
# time and coordinates dimensions
time_dim = 'time'
lon_dim = 'longitude'
let_dim = 'latitude'
z_dim = 'depth'
temp_var = 'thetao'
```

```python
#Choise of z and pcm features (very important to make a good choise) 
z = np.arange(-1.,-2000,-10.)
pcm_features = {'temperature': z}
#classes
K=8
#fit
features_in_ds = {'temperature': temp_var}
```

```python
ds = xr.open_dataset(file_path)
ds['depth'] = -np.abs(ds['depth'].values)
ds.depth.attrs['axis'] = 'Z'
```

```python
z = ds['depth'][0:20]
pcm_features = {'temperature': z}
m = pcm(K=K, features=pcm_features)
features_in_ds = {'temperature': temp_var}
features_zdim=z_dim
m.fit(ds, features=features_in_ds, dim=features_zdim)
m.to_netcdf('model_new.nc')
```

```python
#Load model and create new pcm
m2 = pyxpcm.load_netcdf('model_new.nc')
# Reload data (just to be sure)
ds2 = xr.open_dataset(file_path)
ds2['depth'] = -ds2['depth']
ds2.depth.attrs['axis'] = 'Z'
print(m2)
```

```python
m2.predict(ds2,features = features_in_ds, dim = features_zdim, inplace=True)
```
